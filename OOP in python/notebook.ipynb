# Two data structure

# Series
* one-dimensional labeled array
* similar to a python list
* stores one column of data

```python
import pandas as pd

data = pd.Series([10,20,30,40])
print(data)
```

# Dataframe
* two-dimensional table
* rows and column 
* most commonly used pandas object


```python

data = {
    "Name": ["Ram", "Hari", "Sita"],
    "Age": [20, 21, 22],
    "City": ["Pokhara", "Kathmandu", "Butwal"]
}

df = pd.DataFrame(data)

print(df)
```

# Clean Data
* Check for missing values
>print(df.isnull().sum())
* Remove rows with missing values
>df = df.dropna()
* Fill missing values
>df = df.fillna(0)

or

>df["Marks"] = df["Marks"].fillna(df["Marks"].mean())
* Remove duplicate rows
>df = df.drop_duplicates()


```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Department"] = df["Department"].fillna("Unknown")
df["GPA"] = df["GPA"].fillna(df["GPA"].mean())
print(df["Age"].mean())
print(df["GPA"].mean())
print(df)
```

# Analyze Data
* Find average marks
>print(df["Marks"].mean())
* Find maximum marks
>print(df["Marks"].max())
* Find minimum marks
>print(df["Marks"].min())
* Count students
>print(df["Name"].count())
* Sort by marks
>print(df.sort_values(by="Marks", ascending=False))
* Filter students with marks greater than 80
>print(df[df["Marks"] > 80])
* Group by city
>print(df.groupby("City")["Marks"].mean())

```python

# Analyze
print("\nAverage GPA:", df["GPA"].mean())
print("Highest GPA:", df["GPA"].max())
print("Average Age:", df["Age"].mean())

print("\nStudents with GPA > 3.7")
print(df[df["GPA"] > 3.7])

print("\nStudents by Gender")
print(df["Gender"].value_counts())

print("\nAverage GPA by Department")
print(df.groupby("Department")["GPA"].mean())

# Save cleaned data
df.to_csv("students_cleaned.csv", index=False)
```
